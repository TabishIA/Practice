```python
'''
Regression: Simple Linear 
Dataset: 1.01. Simple linear regression 
Apply Data pre-processing (Label Encoding , Data Transformation….) techniques if necessary. 
Explore the relationship between students SAT score and GPA using linear regression model. 
Also display the regression results and plot the regression line. 
'''
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

```python
data = pd.read_csv(r'LP-1_Dataset\Regression\1.01. Simple linear regression.csv')

data.head()
```

```python
data.isnull().sum()
```

```python
X = data[['SAT']]
y = data['GPA']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
slope = model.coef_[0]
coefficient = model.coef_
intercept = model.intercept_
rss = np.sum((y_test - y_pred) ** 2)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

n = len(y_test)
p = 1
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print("Slope: ", slope)
print("Coefficient: ", coefficient)
print("Intercept: ", intercept)
print("RSS: ", rss)
print("MAE: ", mae)
print("MSE: ", mse)
print("RMSE: ", rmse)
print("R^2: ", r2)
print("Adjusted R^2: ", adjusted_r2)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label="Actual Data")
plt.plot(X, model.predict(X), color='red', label="Regression Line")
plt.xlabel("SAT")
plt.ylabel("GPA")
plt.title("SAT vs. GPA")
plt.legend()
plt.show()
```
