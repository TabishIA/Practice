```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'LP-1_Dataset\Regression\diabetes.csv')

data.head()
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
data.dtypes
```

```python
X = data[['Glucose']]
y = data['Outcome']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# calculate the coefficients
coefficients = model.coef_
intercept = model.intercept_
```

```python
# calculate residual sum of square (RSS)
rss = np.sum((y_test - y_pred) ** 2)
```

```python
# calculate the coefficient of determination (R^2 Score)
r2 = r2_score(y_test, y_pred)
```

```python
print("Coefficients: ", coefficients)
print("Intercept: ", intercept)
print("Residual Sum of Squares (RSS): ", rss)
print("Coefficient of Determination (R^2): ", r2)
```

```python
# mean absolute error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# mean squared error (MSE)
mse = mean_squared_error(y_test, y_pred)

# root mean squared error (RMSE)
rmse = np.sqrt(mse)

# r-squared (coefficient of determination)
r2 = r2_score(y_test, y_pred)

# adjusted r-squared
n = len(y_test) # number of data points
p = 1 # number of predictors (assuming 1 features in this case)
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
print(f'R^2: {r2}')
print(f'Adjusted R^2: {adjusted_r2}')
```

```python
# relationship between independent variable and the dependent variable
plt.figure(figsize=(10, 6))
plt.scatter(X, y)
plt.xlabel('Feature/Glucose')
plt.ylabel('Target/Diabetes')
plt.title('Glucose vs. Diabetes')
plt.show()
```

```python
# how regression line fits the data
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.xlabel('Feature/Glucose')
plt.ylabel('Target/Diabetes')
plt.title('Fitted Regression Line')
plt.show()
```

```python
data = pd.read_csv(r'LP-1_Dataset\Regression\diabetes.csv')

data.head()
```

```python
X = data[['Glucose']]
y = data['BMI']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
coefficients = model.coef_
intercepts = model.intercept_
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rss = np.sum((y_test - y_pred) ** 2)
rmse = np.sqrt(mse)

n = len(y_test)
p = 1
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print("Coefficients: ", coefficients)
print("Intercepts: ", intercepts)
print("R^2: ", r2)
print("MAE: ", mae)
print("MSE: ", mse)
print("RMSE: ", rmse)
print("RSS: ", rss)
print("Adjusted R^2: ", adjusted_r2)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y)
plt.xlabel("BMI")
plt.ylabel("Glucose")
plt.title("Glucose vs. BMI")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.xlabel("BMI")
plt.ylabel("Glucose")
plt.title("Fitted Regression Line")
plt.show()
```
